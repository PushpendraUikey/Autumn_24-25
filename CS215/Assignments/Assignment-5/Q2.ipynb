```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the refined dataset

```python
df = pd.read_csv("/kaggle/input/cleaned-data/cleaned_taxi_data.csv")
```

### 2.a

#### Trip Frequency by Hour

```python
# Get the count of trips per hour
hour_wise_trip = df['trip_start_hour'].value_counts().reindex(range(24), fill_value=0)

# Convert the result to a DataFrame with appropriate column names

hour_wise_trip = hour_wise_trip.reset_index()
hour_wise_trip.columns = ['trip_start_hour', 'trip_count']
# print(hour_wise_trip[:])

# plotting the trip_hours vs no of trips
plt.figure(figsize=(10, 5))
plt.bar(hour_wise_trip['trip_start_hour'], hour_wise_trip['trip_count'], color='green')
plt.xlabel('Trip Start Hour')
plt.ylabel('Number of Trips')
plt.title('Number of Trips by Start Hour')
plt.xticks(range(24))  # Ensures all hours (0-23) are labeled on the x-axis
plt.show()
```

#### Trip Frequency by Day

```python
# Get the count of Trip Start Day
day_wise_trip = df['trip_start_day'].value_counts().reindex(range(1,8,1), fill_value=0)

# Convert the result to a DataFrame with appropriate column names
day_wise_trip = day_wise_trip.reset_index()
day_wise_trip.columns = ['trip_start_day', 'trip_count']
# print(day_wise_trip[:])

# plotting the trip_day vs no of trips
plt.figure(figsize=(10, 5))
plt.bar(day_wise_trip['trip_start_day'], day_wise_trip['trip_count'], color='orange')
plt.xlabel('Trip Start Day')
plt.ylabel('Number of Trips')
plt.title('Number of Trips by Start Day')
plt.xticks(range(1,8,1))  # Ensures all hours (0-23) are labeled on the x-axis
plt.show()
```

#### Trip frequency by months

```python
# Get the count of Trip Start Month
month_wise_trip = df['trip_start_month'].value_counts().reindex(range(1,13,1), fill_value=0)

# Convert the result to a DataFrame with appropriate column names
month_wise_trip = month_wise_trip.reset_index()
month_wise_trip.columns = ['trip_start_month', 'trip_count']
# print(month_wise_trip[:])

# plotting the trip_hours vs no of trips
plt.figure(figsize=(10, 5))
plt.bar(month_wise_trip['trip_start_month'], month_wise_trip['trip_count'], color='violet')
plt.xlabel('Trip Start Month')
plt.ylabel('Number of Trips')
plt.title('Number of Trips by Start Month')
plt.xticks(range(1,13,1))  # Ensures all hours (1-12) are labeled on the x-axis
plt.show()
```

#### peak hours are: 18-20
#### peak days are: Saturday and Sunday
#### peak months are: May, October

### 2.b

### HeatMap showing trip frequency by the hour

```python
"""  Heat Map  """
# Reshape the data to a 1x24 matrix for the heatmap
hour_wise_trip_matrix = np.array(hour_wise_trip['trip_count']).reshape(1, 24)

# Plot the heatmap
plt.figure(figsize=(12, 2))
sns.heatmap(hour_wise_trip_matrix, cmap="YlGnBu", cbar=True, annot=False, fmt="d")
plt.xlabel("Hour of the Day")
plt.yticks([])  # Hide y-axis as there's only one row
plt.title("Trip Frequency by Hour")
plt.xticks(ticks=np.arange(24) + 0.5, labels=range(24))  # Center x-axis labels

plt.show()
```

#### correlation analysis

```python
# Filter the data as per your requirement
trip_duration_corr = df.loc[(df['trip_miles'] > 0) & (df['trip_seconds'] > 0), 'trip_seconds']
trip_distance_corr = df.loc[(df['trip_miles'] > 0) & (df['trip_seconds'] > 0), 'trip_miles']

# Calculate the correlation between trip duration and trip distance
correlation = trip_duration_corr.corr(trip_distance_corr)
print("Correlation between Trip Duration and Trip Distance:", correlation)

# Scatter plot of trip duration vs trip distance

plt.figure(figsize=(10, 6))
plt.scatter(trip_distance_corr, trip_duration_corr, alpha=1, color='blue')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Trip Duration (seconds)')
plt.title('Scatter Plot of Trip Duration vs Trip Distance')
plt.show()
```

Outliers: This graph has some points that stand out as outliers: in particular, a trip of 1,710 miles distance, but others might be just an incredibly long duration 72,120 seconds, and so on with a duration above 10,000 seconds.

Clustered data: In general, most of the trips were clustered with streams short in distance (less than 250 miles) and also time streams (less than 10,000 seconds), indicating that, on average, trips in this dataset were short in terms of time and distance, most probably related to local or urban travel.

Route Planning: It is the clustered data, which contains a lower distance and duration; therefore, from the route optimization perspective, it will offer a better predictable model. Outliers will cause distorted predictions since they inflate the average; however, exclusion may result in predictions that better match actual trip characteristics.

### 2.d

```python
monthly_avg_distance = df.groupby('trip_start_month')['trip_miles'].mean()

# Plotting the number of trips and average trip distance on two line charts
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot the number of trips
ax1.set_xlabel('Month')
ax1.set_ylabel('Number of Trips', color='blue')
ax1.plot(month_wise_trip['trip_start_month'], month_wise_trip['trip_count'], color='blue', marker='o', label='Number of Trips')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis to plot the average trip distance
ax2 = ax1.twinx()
ax2.set_ylabel('Average Trip Distance (miles)', color='green')
ax2.plot(monthly_avg_distance.index, monthly_avg_distance.values, color='green', marker='s', label='Average Trip Distance')
ax2.tick_params(axis='y', labelcolor='green')

# Add titles and grid
plt.title('Monthly Trends: Number of Trips and Average Trip Distance')
ax1.grid(True)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

1. **High Trip Frequency in Peak Months**  
   Observing spikes in certain months suggests increased demand for taxis during these times, such as on weekends or in summer months.

   **Resource Allocation Implications**
   - **Driver Availability**: Increasing the number of available drivers can help manage the heightened demand during peak hours in these months.
   - **Optimized Dispatching**: To minimize wait times, taxi services should prioritize efficient dispatching, focusing on high-demand areas like airports, shopping malls, or tourist hotspots.

2. **Influence of Seasonal Events**  
   Some events lead to notable increases in both the number of trips and average trip distances.

   **Resource Allocation Implications**  
   - **Event-Based Scheduling**: Taxi services can anticipate these spikes by expanding fleet availability around major events or peak traffic periods.
   - **Surge Pricing**: Applying surge pricing during high-demand events can help balance demand, motivating more drivers to be on the road.
   - **Targeted Marketing**: To capture event-driven demand, taxi companies can engage in focused marketing campaigns in areas with high traffic from events.

3. **Recognizing Low-Activity Periods**  
   Certain months, such as those following holidays or off-peak seasons, may show lower trip frequencies, signaling reduced taxi demand.

   **Resource Allocation Implications**  
   - **Reduced Fleet Size**: During low-demand months, limiting the active fleet can prevent oversupply and reduce costs.
   - **Seasonal Staffing Adjustments**: With fewer trips during these periods, adjusting driver schedules or reducing hours can help optimize labor expenses.

### 2.e

```python
# Group the data by 'trip_start_hour' and calculate the average trip distance for each hour
# Filter out rows where trip_miles is zero
df_filtered = df[df['trip_miles'] > 0]
# print(df_filtered.shape)
# Now group by 'trip_start_hour' and calculate the mean of trip_miles
hour_wise_distance = df_filtered.groupby('trip_start_hour')['trip_miles'].mean()
# print(hour_wise_distance[:])


# Plotting the variation in trip distances across hours of the day
plt.figure(figsize=(10, 6))
hour_wise_distance.plot(marker='o', color='b', linestyle='-', markersize=6)

# Adding titles and labels for clarity
plt.title('Variation in Trip Distances Across Hours of the Day', fontsize=16)
plt.xlabel('Hour of the Day', fontsize=14)
plt.ylabel('Average Trip Distance (miles)', fontsize=14)
plt.xticks(range(0, 24))  # Ensures ticks for all 24 hours
plt.grid(True)
plt.tight_layout()

# Displaying the plot
plt.show()
```

### Distance vs. Time of Day Plot Interpretation:

#### **1. Peak Hour (Hour 5 AM)**:

- **Peak Hour**: The data shows that **hour 5 AM** is the peak hour having the largest mean distance travelled to (10.16 miles).
- **Insights**:
- The longer distances that need to be traveled during early morning hours would indicate people going to the airport, catching an early flight, or other longer trips for reasons.
- There would be lesser demand for taxis during those times, but the higher distance trips would be farther, out-of-town trips, and even airport travel.
- It's also that maybe people have to travel longer distances, but fewer numbers exist on the road, which might contribute to the relatively higher average distance.

### 2. Off-Peak Hour (Hour 7 PM)
- Off-Peak Hour: The hour 7 PM shows a relatively lower mean trip distance of about 4.35 miles.
- Insights:
While 7 PM could be considered a peak hour for most services, by virtue of evening traffic, average trip distance appears to be lower in this scenario. This could also be because, after mid-night, everybody goes for short trips, at local levels, such as travel between home and restaurants or entertainment venues.
- The sheer number of trips notwithstanding, it is possible that the trip distances are more likely to be bunched up in cities because most of the travel would then be within the city itself and hence shorter distances on average

#### **3. General Trend**:
- **Overall Pattern**: From after 5 AM, average trip distances generally tend to be between **3 and 4 miles** the rest of the day, suggesting that most trips are short-distance. 
-The distance is relatively higher during the late night and early morning periods either because of the longer trip or lesser traffic.
The trend that appears from the graph is that the trips undertaken in the peak hours are shorter in terms of distance. This may also be due to direct activity within the region and congestion on the roads.
 
#### **4. Taxi Service Implications**:
- **Peak Hours (5 AM)** Since hour 5 AM is the peak hour having the highest trip distances, this is to mean that taxi services should be prepared in making longer trips in these hours. This can be out to an airport or inter-town traveling.
  - **Resource Allocation**: The taxi service must have available vehicles to enable longer trips, especially when travelers are going to the airport or when it's out-town travels.
- **Off Peak Hours (7 PM)**: Although 7 PM marks a peak of trip volume, short distances ensure that taxi services can have more frequent local trips. This may include urban concentration, for instance to dinner or events.
- **Operational Changes**: The demand calls for more taxis, but optimal routing and dispatch would possibly be encouraged in order to fit more trips into the evening peak, albeit ones that are shorter than average.
